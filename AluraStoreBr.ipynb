### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja['Loja Origem'] = '1'
loja2['Loja Origem'] = '2'
loja3['Loja Origem'] = '3'
loja4['Loja Origem'] = '4'

todas_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

todas_lojas.head()
```

```python
loja.describe()
```

```python
df = pd.DataFrame(todas_lojas)
```

#1. Análise do faturamento


```python
df.groupby('Loja Origem')['Preço'].sum()
```

# 2. Vendas por Categoria


```python
def agrupar_por_categoria(df):
    coluna_categoria = df.columns[df.columns.str.contains('categoria', case=False)][0]
    agrupado = df.groupby([coluna_categoria, 'Loja Origem']).sum(numeric_only=True)
    return agrupado
```

```python
agrupar_por_categoria(df)['Preço']
```

# 3. Média de Avaliação das Lojas

```python
df.groupby('Loja Origem')['Avaliação da compra'].mean()
```

# 4. Produtos Mais e Menos Vendidos

```python
def agrupar_por_produto(df):
    coluna_produto_mais_vendido = df.columns[df.columns.str.contains('produto', case=False)][0]
    coluna_produto_menos_vendido = df.columns[df.columns.str.contains('produto', case=False)][1]
    agrupado = df.groupby([coluna_produto_mais_vendido, coluna_produto_menos_vendido, 'Loja Origem']).sum(numeric_only=True)
    return agrupado
```

```python
agrupar_por_produto(df)['Preço']
```

# 5. Frete Médio por Loja


```python
df.groupby('Loja Origem')['Frete'].mean()
```
