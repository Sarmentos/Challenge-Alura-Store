### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
print(faturamento_loja)
print(faturamento_loja2)
print(faturamento_loja3)
print(faturamento_loja4)
```

# 2. Vendas por Categoria


```python
agrupado = loja.groupby(loja.columns[loja.columns.str.contains('categoria', case=False)][0]).sum()
agrupado2 = loja2.groupby(loja2.columns[loja2.columns.str.contains('categoria', case=False)][0]).sum()
agrupado3 = loja3.groupby(loja3.columns[loja3.columns.str.contains('categoria', case=False)][0]).sum()
agrupado4 = loja4.groupby(loja4.columns[loja4.columns.str.contains('categoria', case=False)][0]).sum()
```

```python
print(agrupado)
print(agrupado2)
print(agrupado3)
print(agrupado4)
```

# 3. Média de Avaliação das Lojas

```python
media = loja['Avaliação da compra'].mean()
media2 = loja2['Avaliação da compra'].mean()
media3 = loja3['Avaliação da compra'].mean()
media4 = loja4['Avaliação da compra'].mean()
```

```python
print(media)
print(media2)
print(media3)
print(media4)
```

# 4. Produtos Mais e Menos Vendidos

```python
produto = loja.groupby(loja.columns[loja.columns.str.contains('Produto')][0]).sum()
produto2 = loja2.groupby(loja2.columns[loja2.columns.str.contains('Produto')][0]).sum()
produto3 = loja3.groupby(loja3.columns[loja3.columns.str.contains('Produto')][0]).sum()
produto4 = loja4.groupby(loja4.columns[loja4.columns.str.contains('Produto')][0]).sum()
```

```python
print(produto)
print(produto2)
print(produto3)
print(produto4)
```

```python
print(frete_medio)
print(frete_medio2)
print(frete_medio3)
print(frete_medio4)
```
